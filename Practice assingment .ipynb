# Joint plot 
#### A way of understanding the relationship between two variables and the distribution of individuals of each variable

```python
sns.jointplot(x='Age',y='LungCap', data=data,kind='reg')
plt.xlabel('Age')
plt.ylabel('Lungcp')

plt.show()
```

#### Here we have regression line that best fits data points. From the plot we can observe that people of age 15-20 years has more Lung capacity.

# Scattered Plot 
#### A type of data visualization that shows the relationship between different variables. This data is shown by placing various data points between an x- and y-axis. Essentially, each of these data points looks “scattered” around the graph, giving this type of data visualization its name.

###### Refered from :https://visme.co/blog/scatter-plot/#:~:text=A%20scatter%20plot%20is%20a,of%20data%20visualization%20its%20name.

```python
sns.scatterplot(x='LungCap',data=data,y='Height')
plt.xlabel('LungCap')
plt.ylabel('Hight')
plt.title('Hight VS Lungcap')
plt.show()
```

#### scatter plot was shown to represent the lung capacities of the people by their height. Lung closing capacity increases with height.

# Pair plot 
#### A pair plot is a useful tool for determining the optimal combination of features that can be used to explain the relationship between two variables or to identify distinct clusters within a dataset. In addition, it can be used to create basic classification models by drawing simple lines or making linear separations in the data.

```python
sns.pairplot(data)
plt.show()
```

#### The pairs plot builds on two basic figures, the histogram and the scatter plot. The histogram on the diagonal allows us to see the distribution of a single variable while the scatter plot shows the relationship (or lack thereof) between two variables. The default pairs plot by itself often gives us valuable insights. From the histograms, we learn that the height,age and lungcapacity variables are left-skewed. We see that lung capacity and age are positively correlated.

# Pair Plot Based on Gender 

```python
sns.pairplot(data,hue='Gender',palette='seismic')
plt.show()
```

#### From the above pair plot we identify the following
1. The lung capacity of female is higher compared to men 
2. The lung capacity and both female and male increase along with age and male with higher age have higher lung capacity compared to female 
3. As the highest increase the lung capacity also increases for both male and female.
4. Male are taller than female 

#### A pair plot is a valuable tool for determining the optimal features to explain the relationship between two variables or to identify well-defined clusters within a given data set. By plotting various combinations of variables, this tool enables the detection of patterns and potential linear relationships. Moreover, it can facilitate the development of simple classification models by drawing clear boundaries or lines to separate the data into distinct categories.

# Pair Plot Based on Smoke 

```python
sns.pairplot(data, hue= 'Smoke', diag_kind='hist')
plt.show()
```

#### The pair plot analysis reveals several notable findings, including:

1. A higher proportion of individuals in the dataset are non-smokers.

2. Smokers have, on average, lower lung capacity than non-smokers.

3. Age is positively associated with smoking, as older individuals in the dataset have higher rates of smoking.

# Heat Map
#### Heat map analysis is a methodical approach to studying and evaluating heat map data in order to gain insights into user behavior and interaction with a given product or service. By examining this data, it is possible to identify patterns and trends that can inform decisions about how to optimize the user experience. As a result, this technique can be a valuable tool for improving product design and driving business success.

```python
sns.heatmap(data.corr(),annot=True,vmin=-1)
plt.show()
```

#### A heat map is a powerful data visualization tool that utilizes colors to represent values within a two-dimensional context. By providing an immediate and intuitive visual summary of complex information, heat maps can be extremely useful in aiding data analysis and decision-making. In this particular instance, a heat map was generated based on the correlation between variables, with only those containing numeric values being included. Notably, the heat map reveals a strong correlation between lung capacity and height of 0.91, highlighting a potentially important relationship between these two variables.
